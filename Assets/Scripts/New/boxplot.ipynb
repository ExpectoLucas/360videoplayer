```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import erf, sqrt
import os, re
from IPython.display import display


DATA_DIR = "C:/Users/96412/Downloads"
FILE_C = os.path.join(DATA_DIR, "Thesis Experiment - Post-Experience Questionnaire (C).csv")
FILE_G = os.path.join(DATA_DIR, "Thesis Experiment - Post-Experience Questionnaire (G).csv")
FILE_QUIZ = os.path.join(DATA_DIR, "Thesis Experiment - Quiz.csv")

# ---- Colors ----
BLUE   = "#1f77b4"   # Control
ORANGE = "#ff7f0e"   # Guidance

# ---- Helpers ----
def ensure_numeric(df, cols):
    df = df.copy()
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def find_group_col(df):
    
    cand = [c for c in df.columns if c.lower() == "group"]
    if cand:
        return cand[0]
    tail = df.columns[-1]
    if "group" in tail.lower():
        return tail
    
    anyc = [c for c in df.columns if "group" in c.lower()]
    return anyc[0] if anyc else None

def scale_of_item(i:int) -> str:
    
    return "1–7" if 1 <= i <= 25 else "1–5"

# ---- Load ----
df_c_raw = pd.read_csv(FILE_C)
df_g_raw = pd.read_csv(FILE_G)
df_quiz  = pd.read_csv(FILE_QUIZ)

# ---- Reverse-coding by exact TEXT (safer than numbers) ----
REV_BY_TEXT = {
    # TLX performance (so higher composite = higher workload)
    "How satisfied are you with your performance in this task? (Do you feel you viewed and understood the content well?)": 7,
    # IPQ negatives
    "I felt like I was just perceiving pictures rather than being in a real place.": 7,
    "I did not feel present in the virtual space.": 7,
    # Enjoyment negative (boring)
    "I thought this experience was boring.": 7,
    # FOMO confidence (reverse)
    "I think I saw all the important things that happened in the video.": 5,
    # Usability discomfort (reverse)
    "I experienced discomfort (e.g. dizziness or motion sickness) while watching.": 5,
    # Spatial coverage reversed parts
    "I feel that I was able to explore the entire 360° scene and not just one part of it.": 5,
    # Guidance distractor (reverse)
    "The guidance cues were distracting or got in the way of my viewing experience.": 5,
}

def apply_reverse_by_text(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    for col in df.columns:
        if col in REV_BY_TEXT:
            mx = REV_BY_TEXT[col]
            out[col] = pd.to_numeric(out[col], errors="coerce")
            out[col] = (mx + 1) - out[col]
    return out

df_c = apply_reverse_by_text(df_c_raw)
df_g = apply_reverse_by_text(df_g_raw)

# ---- Build group-specific item->column maps (skip first two columns if they are timestamp/email-like) ----
def build_map(df: pd.DataFrame) -> dict:
    m = {}
    # 假设第 1~2 列不是题目（Timestamp/Email），题号从第 3 列起
    for i in range(1, 45):      # items 1..44
        idx = i + 1
        if idx < len(df.columns):
            m[i] = df.columns[idx]
    return m

item_to_col_C = build_map(df_c)
item_to_col_G = build_map(df_g)

# ---- Your latest construct ranges ----
# 1–7
constructs_7 = {
    "Cognitive Load (1–7)": list(range(1, 7)),
    "Presence/Immersion (1–7)": list(range(7, 13)),
    "Involvement/Engagement (1–7)": list(range(13, 17)),
    "Experienced Realism (1–7)": list(range(17, 21)),
    "Enjoyment (1–7)": [21,22,23],      # you said: Enjoyment 21–23
    "Time Perception (1–7)": [24,25],   # you said: Time 24–25
}
# 1–5
constructs_5 = {
    "FOMO (1–5)": [26,27,28,29],                      # you said: 26–29 (29 reversed by TEXT)
    "Navigation & Usability (1–5)": [30,31,32,33,35], # you said: includes 35
    "Spatial Coverage & Missed Content (1–5)": [36,37,38],
    "Guidance Usability & Usefulness (1–5, G-only)": [39,40,41,42,43,44],
}

# ---- Print listings for verification (Item, Scale, Reverse?, Text) ----
def print_construct_listing(df_ref, item_map, constructs, title_prefix=""):
    print(f"===== {title_prefix} 构念题目清单（按该表列名） =====")
    for cname, items in constructs.items():
        rows = []
        for i in items:
            col = item_map.get(i, None)
            text = col if col is not None else f"[缺列 i={i}]"
            rows.append({
                "Item": i,
                "Scale": scale_of_item(i),
                "Reverse?": "Yes" if (col in REV_BY_TEXT) else "No",
                "Text": text
            })
        print(f"\n— {cname}")
        display(pd.DataFrame(rows, columns=["Item","Scale","Reverse?","Text"]))

print_construct_listing(df_c, item_to_col_C, constructs_7, title_prefix="Control (1–7)")
print_construct_listing(df_c, item_to_col_C, constructs_5, title_prefix="Control (1–5)")
print_construct_listing(df_g, item_to_col_G, constructs_7, title_prefix="Guidance (1–7)")
print_construct_listing(df_g, item_to_col_G, constructs_5, title_prefix="Guidance (1–5)")

# ---- Compute construct scores (mean per participant, skip NaNs), keep SubjectKey ----
def scores_from_items(df, item_map, items):
    cols = [item_map[i] for i in items if i in item_map]
    if not cols:
        return pd.Series([], dtype=float)
    tmp = ensure_numeric(df, cols)
    return tmp[cols].mean(axis=1, skipna=True)

def build_scores(df, item_map, constructs):
    out = {}
    for name, items in constructs.items():
        out[name] = scores_from_items(df, item_map, items)
    return pd.DataFrame(out)

# locate subject key column
key_c = find_group_col(df_c)
key_g = find_group_col(df_g)
if key_c is None or key_g is None:
    raise RuntimeError("找不到用于连接的 group 键，请确认各表最后一列名里包含 'group' / 'Group'。")

control  = pd.concat([build_scores(df_c, item_to_col_C, constructs_7),
                      build_scores(df_c, item_to_col_C, constructs_5)], axis=1)
control["GroupLabel"] = "Control"
control["SubjectKey"] = df_c[key_c].astype(str)

guidance = pd.concat([build_scores(df_g, item_to_col_G, constructs_7),
                      build_scores(df_g, item_to_col_G, constructs_5)], axis=1)
guidance["GroupLabel"] = "Guidance"
guidance["SubjectKey"] = df_g[key_g].astype(str)

all_scores = pd.concat([control, guidance], ignore_index=True)

print("\n===== 构念得分（示例前5行） =====")
display(all_scores.head())

# ---- Prepare Quiz (parse % and keep Group as key) ----
def parse_score_cell(x):
    if isinstance(x, str) and "/" in x:
        try:
            a,b = x.split("/")
            a=float(a.strip()); b=float(b.strip())
            return a, b, a/b*100.0
        except:
            return np.nan, np.nan, np.nan
    return np.nan, np.nan, np.nan

quiz = df_quiz.copy()
nums = quiz["Score"].apply(parse_score_cell)
quiz["Score_%"] = [t[2] for t in nums]
for col in ["Score(without POI)", "Score(POI)"]:
    quiz[col] = pd.to_numeric(quiz[col], errors="coerce")

# ---- JOIN: all_scores.SubjectKey  <->  quiz.Group ----
if "Group" not in quiz.columns:
    # 有些导出会把列名小写或放在最后一列
    q_key = find_group_col(quiz)
    if q_key is None:
        raise RuntimeError("Quiz 里也需要一个 'Group' 键列（例如 Guidance_1）。没找到。")
    quiz = quiz.rename(columns={q_key: "Group"})

merged = pd.merge(
    all_scores, 
    quiz[["Group","Score_%","Score(without POI)","Score(POI)"]],
    left_on="SubjectKey", right_on="Group", how="left"
)

print("\n===== Join 状态检查 =====")
n_total = len(all_scores)
n_matched = merged["Score_%"].notna().sum()
print(f"总样本: {n_total}，与 Quiz 成功匹配: {n_matched}，未匹配: {n_total - n_matched}")
if n_total - n_matched > 0:
    print("未匹配样例（最多 5 条）：")
    display(merged.loc[merged["Score_%"].isna(), ["SubjectKey","GroupLabel"]].head())

print("\n===== 合并后的数据（前 5 行）=====")
display(merged.head())

# ---- Welch t-tests (between GroupLabel) ----
def welch(a, b):
    a = pd.to_numeric(pd.Series(a), errors="coerce").dropna().values
    b = pd.to_numeric(pd.Series(b), errors="coerce").dropna().values
    if len(a) < 2 or len(b) < 2:
        return np.nan, np.nan
    m1, m2 = a.mean(), b.mean()
    s1, s2 = a.var(ddof=1), b.var(ddof=1)
    t = (m1 - m2) / np.sqrt(s1/len(a) + s2/len(b))
    try:
        from scipy.stats import t as tdist
        df = (s1/len(a) + s2/len(b))**2 / ( (s1**2)/((len(a)**2)*(len(a)-1)) + (s2**2)/((len(b)**2)*(len(b)-1)) )
        p = 2 * tdist.sf(abs(t), df)
    except Exception:
        p = 2 * (1 - 0.5*(1+erf(abs(t)/sqrt(2))))
    return float(t), float(p)

rows = []
construct_cols = [c for c in all_scores.columns if c not in ["GroupLabel","SubjectKey"]]
for name in construct_cols:
    a = merged.loc[merged["GroupLabel"]=="Control", name]
    b = merged.loc[merged["GroupLabel"]=="Guidance", name]
    t, p = welch(a, b)
    rows.append({
        "Construct": name,
        "Control_Mean": float(pd.to_numeric(a, errors="coerce").mean()),
        "Guidance_Mean": float(pd.to_numeric(b, errors="coerce").mean()),
        "t_Welch": t,
        "p_value": p,
        "n_control": int(pd.to_numeric(a, errors="coerce").notna().sum()),
        "n_guidance": int(pd.to_numeric(b, errors="coerce").notna().sum()),
    })
stats_df = pd.DataFrame(rows)
print("\n===== Welch t 结果 =====")
display(stats_df.sort_values("Construct"))

# ---- Correlations with Quiz outcomes (pooled) ----
quiz_targets = ["Score_%","Score(POI)","Score(without POI)"]
corr_rows = []
for qcol in quiz_targets:
    for ccol in construct_cols:
        x = merged[ccol]
        y = merged[qcol]
        x = pd.to_numeric(x, errors="coerce")
        y = pd.to_numeric(y, errors="coerce")
        valid = x.notna() & y.notna()
        if valid.sum() >= 3:
            r = np.corrcoef(x[valid], y[valid])[0,1]
        else:
            r = np.nan
        corr_rows.append({"QuizVar": qcol, "Construct": ccol, "r_Pearson": r, "n": int(valid.sum())})
corr_df = pd.DataFrame(corr_rows)
print("\n===== 构念 与 Quiz 的皮尔逊相关（合并全样本）=====")
display(corr_df.pivot(index="Construct", columns="QuizVar", values="r_Pearson").round(3))



# ======== Append: print an English explanation under every boxplot ========
import numpy as np
import pandas as pd

def _describe_group(series):
    s = pd.to_numeric(series, errors="coerce").dropna()
    return {
        "n": int(s.size),
        "mean": float(s.mean()) if s.size else np.nan,
        "median": float(s.median()) if s.size else np.nan,
        "sd": float(s.std(ddof=1)) if s.size > 1 else np.nan,
    }

def _scale_hint(ylabel, col):
    y = (ylabel or "") + " " + (col or "")
    ylow = y.lower()
    # try to infer the scale from axis label or column name
    if ("1–7" in ylabel) or ("1-7" in ylabel):
        return " (scale: 1–7)"
    if ("1–5" in ylabel) or ("1-5" in ylabel):
        return " (scale: 1–5)"
    if "percent" in ylow or "%" in ylow:
        return " (%)"
    if "score" in ylow:
        return " (range: 0–11)"
    return ""

# If 'welch' is already defined above, we'll reuse it. Otherwise, define a fallback.
try:
    welch
except NameError:
    from math import erf, sqrt
    def welch(a, b):
        a = pd.to_numeric(pd.Series(a), errors="coerce").dropna().values
        b = pd.to_numeric(pd.Series(b), errors="coerce").dropna().values
        if len(a) < 2 or len(b) < 2:
            return np.nan, np.nan
        m1, m2 = a.mean(), b.mean()
        s1, s2 = a.var(ddof=1), b.var(ddof=1)
        t = (m1 - m2) / np.sqrt(s1/len(a) + s2/len(b))
        # normal approx for p-value (two-tailed) if SciPy isn't available
        p = 2 * (1 - 0.5*(1+erf(abs(t)/sqrt(2))))
        return float(t), float(p)

def boxplot_two_groups_explained(df, col, title, ylabel, ylim=None):
    # Data
    cvals = pd.to_numeric(df.loc[df["GroupLabel"]=="Control",  col], errors="coerce").dropna()
    gvals = pd.to_numeric(df.loc[df["GroupLabel"]=="Guidance", col], errors="coerce").dropna()

    # Plot (keep your original blue/orange style)
    data = [cvals.values, gvals.values]
    labels = ["Control", "Guidance"]
    fig = plt.figure(figsize=(6,4))
    bp = plt.boxplot(data, positions=[1,2], widths=0.6, showmeans=True, patch_artist=True)
    colors = [BLUE, ORANGE]
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color); patch.set_alpha(0.5); patch.set_edgecolor(color)
    for whisker, color in zip(bp['whiskers'], [BLUE, BLUE, ORANGE, ORANGE]): whisker.set_color(color)
    for cap, color in zip(bp['caps'], [BLUE, BLUE, ORANGE, ORANGE]): cap.set_color(color)
    for median, color in zip(bp['medians'], [BLUE, ORANGE]): median.set_color(color); median.set_linewidth(2)
    for mean, color in zip(bp['means'], [BLUE, ORANGE]):
        mean.set_marker('o'); mean.set_markerfacecolor(color); mean.set_markeredgecolor(color)
    plt.xticks([1,2], labels)
    plt.title(title)
    plt.ylabel(ylabel)
    if ylim: plt.ylim(ylim)
    plt.tight_layout()
    plt.show()

    # Compute summary + significance
    cdesc = _describe_group(cvals)
    gdesc = _describe_group(gvals)
    t, p = welch(cvals, gvals)
    diff = (gdesc["mean"] - cdesc["mean"]) if np.isfinite(gdesc["mean"]) and np.isfinite(cdesc["mean"]) else np.nan

    # Print an English explanation under the figure
    hint = _scale_hint(ylabel, col)
    print(f"[Figure note] {title}{hint}")
    print(f"  Control:   n={cdesc['n']}, M={cdesc['mean']:.2f}, Median={cdesc['median']:.2f}, SD={cdesc['sd']:.2f}")
    print(f"  Guidance:  n={gdesc['n']}, M={gdesc['mean']:.2f}, Median={gdesc['median']:.2f}, SD={gdesc['sd']:.2f}")
    print(f"  Mean difference (Guidance − Control) = {diff:.2f}; Welch t = {t:.3f}, p = {p:.3f}")
# ========================================================================



# ---- Boxplots (Blue=Control, Orange=Guidance) ----
def boxplot_two_groups(df, col, title, ylabel, ylim=None):
    cvals = pd.to_numeric(df.loc[df["GroupLabel"]=="Control", col], errors="coerce").dropna().values
    gvals = pd.to_numeric(df.loc[df["GroupLabel"]=="Guidance", col], errors="coerce").dropna().values
    data = [cvals, gvals]
    labels = ["Control", "Guidance"]
    fig = plt.figure(figsize=(6,4))
    bp = plt.boxplot(data, positions=[1,2], widths=0.6, showmeans=True, patch_artist=True)
    colors = [BLUE, ORANGE]
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color); patch.set_alpha(0.5); patch.set_edgecolor(color)
    for whisker, color in zip(bp['whiskers'], [BLUE, BLUE, ORANGE, ORANGE]):
        whisker.set_color(color)
    for cap, color in zip(bp['caps'], [BLUE, BLUE, ORANGE, ORANGE]):
        cap.set_color(color)
    for median, color in zip(bp['medians'], [BLUE, ORANGE]):
        median.set_color(color); median.set_linewidth(2)
    for mean, color in zip(bp['means'], [BLUE, ORANGE]):
        mean.set_marker('o'); mean.set_markerfacecolor(color); mean.set_markeredgecolor(color)
    plt.xticks([1,2], labels)
    plt.title(title)
    plt.ylabel(ylabel)
    if ylim: plt.ylim(ylim)
    plt.tight_layout()
    plt.show()

# 1–7 constructs
for name in ["Cognitive Load (1–7)","Presence/Immersion (1–7)","Involvement/Engagement (1–7)",
             "Experienced Realism (1–7)","Enjoyment (1–7)","Time Perception (1–7)"]:
    boxplot_two_groups_explained(merged, name, name, "Response (1–7)", ylim=(1,7))

# 1–5 constructs
for name in ["FOMO (1–5)","Navigation & Usability (1–5)",
             "Spatial Coverage & Missed Content (1–5)","Guidance Usability & Usefulness (1–5, G-only)"]:
    boxplot_two_groups_explained(merged, name, name, "Response (1–5)", ylim=(1,5))

# Quiz boxplots (optional)
boxplot_two_groups_explained(merged, "Score(without POI)", "Quiz Score Non-POI", "Score")
boxplot_two_groups_explained(merged, "Score(POI)", "Quiz Score POI", "Score")
```

```python
# === New cell: Grouped boxplots by item (x-axis = question ID) for FOMO & Missed Content ===
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import numpy as np
import pandas as pd

# -- helpers (use existing maps/dataframes and reversed-scored df_c/df_g) --
def _item_text(idx:int) -> str:
    col_c = item_to_col_C.get(idx)
    col_g = item_to_col_G.get(idx)
    if isinstance(col_c, str): 
        return col_c
    if isinstance(col_g, str): 
        return col_g
    return f"Item {idx}"

def _get_vals_for_item(idx:int):
    col_c = item_to_col_C.get(idx)
    col_g = item_to_col_G.get(idx)
    cvals = pd.Series([], dtype=float)
    gvals = pd.Series([], dtype=float)
    if (col_c is not None) and (col_c in df_c.columns):
        cvals = pd.to_numeric(df_c[col_c], errors="coerce").dropna()
    if (col_g is not None) and (col_g in df_g.columns):
        gvals = pd.to_numeric(df_g[col_g], errors="coerce").dropna()
    return cvals, gvals

def _style_single_box(bp, color):
    # bp is the dict returned by plt.boxplot when plotting a single series
    bp['boxes'][0].set_facecolor(color)
    bp['boxes'][0].set_alpha(0.5)
    bp['boxes'][0].set_edgecolor(color)
    for w in bp['whiskers']:
        w.set_color(color)
    for c in bp['caps']:
        c.set_color(color)
    bp['medians'][0].set_color(color)
    bp['medians'][0].set_linewidth(2)
    bp['means'][0].set_marker('o')
    bp['means'][0].set_markerfacecolor(color)
    bp['means'][0].set_markeredgecolor(color)

def grouped_boxplot_for_construct(items, title, ylabel="Response (1–5)"):
    """
    Draw one figure with side-by-side boxes per item (Control vs Guidance).
    Also print English notes per item below.
    """
    fig = plt.figure(figsize=(1.8*len(items)+2, 4.6))
    offset = 0.18  # horizontal offset for side-by-side boxes
    base_positions = np.arange(1, len(items)+1)

    # plot per item: control at base-offset, guidance at base+offset
    for j, idx in enumerate(items):
        pos_c = base_positions[j] - offset
        pos_g = base_positions[j] + offset

        cvals, gvals = _get_vals_for_item(idx)

        # Control
        bp_c = plt.boxplot([cvals.values], positions=[pos_c], widths=0.32,
                           showmeans=True, patch_artist=True)
        _style_single_box(bp_c, BLUE)

        # Guidance
        bp_g = plt.boxplot([gvals.values], positions=[pos_g], widths=0.32,
                           showmeans=True, patch_artist=True)
        _style_single_box(bp_g, ORANGE)

    plt.xticks(base_positions, [str(i) for i in items])
    plt.xlabel("Question ID")
    plt.title(title)
    plt.ylabel(ylabel)
    plt.ylim(1, 5)
    plt.xlim(0.5, len(items)+0.5)

    # legend
    handles = [Patch(facecolor=BLUE, alpha=0.5, edgecolor=BLUE, label="Control"),
               Patch(facecolor=ORANGE, alpha=0.5, edgecolor=ORANGE, label="Guidance")]
    plt.legend(handles=handles, loc="upper right", frameon=False)
    plt.tight_layout()
    plt.show()

    # ---- English notes per item ----
    print(f"[Figure note] {title} (scale: 1–5)")
    for idx in items:
        cvals, gvals = _get_vals_for_item(idx)
        cdesc = _describe_group(cvals)
        gdesc = _describe_group(gvals)
        t, p = welch(cvals, gvals)
        diff = (gdesc["mean"] - cdesc["mean"]) if np.isfinite(gdesc["mean"]) and np.isfinite(cdesc["mean"]) else np.nan
        qtext = _item_text(idx)
        print(f"  Q{idx} — {qtext}")
        print(f"    Control:   n={cdesc['n']}, M={cdesc['mean']:.2f}, Median={cdesc['median']:.2f}, SD={cdesc['sd']:.2f}")
        print(f"    Guidance:  n={gdesc['n']}, M={gdesc['mean']:.2f}, Median={gdesc['median']:.2f}, SD={gdesc['sd']:.2f}")
        print(f"    Mean difference (Guidance − Control) = {diff:.2f}; Welch t = {t:.3f}, p = {p:.3f}")

# --- Run for FOMO and Missed Content ---
FOMO_ITEMS = [26, 27, 28, 29]
MISSED_ITEMS = [36, 37, 38]  # Spatial Coverage & Missed Content

grouped_boxplot_for_construct(FOMO_ITEMS, "FOMO items (Q26–Q29)")
grouped_boxplot_for_construct(MISSED_ITEMS, "Missed Content items (Q36–Q38)")
```

```python
# =========================
# Fix: attach Quiz total Score to `merged` and do scatter + r/p/n
# =========================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import erf, sqrt
from IPython.display import display

# --- helpers ---
def find_group_col(df):
    cand = [c for c in df.columns if c.lower() == "group"]
    if cand: return cand[0]
    tail = df.columns[-1]
    if "group" in str(tail).lower(): return tail
    anyc = [c for c in df.columns if "group" in str(c).lower()]
    return anyc[0] if anyc else None

def fisher_z_pvalue(r, n):
    if n is None or n < 4 or not np.isfinite(r): return np.nan
    z = 0.5 * np.log((1 + r) / (1 - r))
    z_se = 1 / np.sqrt(n - 3)
    z_stat = abs(z) / z_se
    return float(2 * (1 - 0.5*(1+erf(z_stat / np.sqrt(2)))))

def pearson_r(x, y):
    x = pd.to_numeric(pd.Series(x), errors="coerce")
    y = pd.to_numeric(pd.Series(y), errors="coerce")
    m = x.notna() & y.notna()
    n = int(m.sum())
    if n < 3: return np.nan, n
    return float(np.corrcoef(x[m], y[m])[0,1]), n

# --- 1) 把 Quiz 的 Score 合并进 merged ---
if 'merged' not in globals():
    raise RuntimeError("没有找到 `merged`。请先运行上一格（构念计算 + Quiz 连接）。")

# 有 df_quiz 就用；没有就尝试用 FILE_QUIZ 重新载入
if 'df_quiz' not in globals():
    if 'FILE_QUIZ' not in globals():
        raise RuntimeError("找不到 df_quiz 或 FILE_QUIZ。请先运行加载 CSV 的单元格。")
    df_quiz = pd.read_csv(FILE_QUIZ)

quiz_key = find_group_col(df_quiz)
if quiz_key is None:
    raise RuntimeError("Quiz CSV 里找不到 group 键列（例如 'Group'）。")

# 把 Score 解析成数值（若是 '8/11' 这种，取前面的 8；若本来就是数值就直接用）
score_series = pd.to_numeric(df_quiz['Score'], errors='coerce')
if score_series.isna().all() and df_quiz['Score'].astype(str).str.contains('/').any():
    score_series = df_quiz['Score'].astype(str).str.split('/').str[0].astype(float)

quiz_for_merge = pd.DataFrame({
    "Group": df_quiz[quiz_key].astype(str),
    "Score": score_series
})

# 如果 merged 里面已经有 Score，就不重复合并；否则合一次
if 'Score' not in merged.columns or merged['Score'].isna().all():
    merged = pd.merge(
        merged,
        quiz_for_merge,
        left_on="SubjectKey", right_on="Group",
        how="left", suffixes=("","_quiz")
    )

print("合并完成：",
      f"总样本={len(merged)}, 成功匹配 Score 的样本={merged['Score'].notna().sum()}")

# --- 2) 只用 Quiz 总分 Score 做散点 + r/p/n ---
score_col = 'Score'
merged[score_col] = pd.to_numeric(merged[score_col], errors='coerce')

construct_cols = [
    "Cognitive Load (1–7)",
    "Presence/Immersion (1–7)",
    "Involvement/Engagement (1–7)",
    "Experienced Realism (1–7)",
    "Enjoyment (1–7)",                    # 21–23
    "Time Perception (1–7)",              # 24–25
    "FOMO (1–5)",                         # 26–29
    "Navigation & Usability (1–5)",       # 30,31,32,33,35
    "Spatial Coverage & Missed Content (1–5)"
]

BLUE   = "#1f77b4"   # Control
ORANGE = "#ff7f0e"   # Guidance

def scatter_and_table(df, xcol, ycol, title, xlabel, ylabel):
    xc = pd.to_numeric(df.loc[df["GroupLabel"]=="Control",  xcol], errors="coerce")
    yc = pd.to_numeric(df.loc[df["GroupLabel"]=="Control",  ycol], errors="coerce")
    xg = pd.to_numeric(df.loc[df["GroupLabel"]=="Guidance", xcol], errors="coerce")
    yg = pd.to_numeric(df.loc[df["GroupLabel"]=="Guidance", ycol], errors="coerce")

    plt.figure(figsize=(6,5))
    if xc.notna().any() and yc.notna().any():
        plt.scatter(xc, yc, s=55, alpha=0.85, edgecolor="none", label="Control", c=BLUE)
    if xg.notna().any() and yg.notna().any():
        plt.scatter(xg, yg, s=55, alpha=0.85, edgecolor="none", label="Guidance", c=ORANGE)

    # overall fit line
    xa = pd.to_numeric(df[xcol], errors="coerce")
    ya = pd.to_numeric(df[ycol], errors="coerce")
    m  = xa.notna() & ya.notna()
    if m.sum() >= 2:
        xs, ys = xa[m].values, ya[m].values
        k, b = np.polyfit(xs, ys, deg=1)
        xx = np.linspace(xs.min(), xs.max(), 100)
        plt.plot(xx, k*xx + b, lw=2, alpha=0.85, color="gray", label="Linear fit (all)")

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(frameon=False)
    plt.tight_layout()
    plt.show()

    # r/p/n 表（All / Control / Guidance）
    r_all, n_all = pearson_r(xa, ya); p_all = fisher_z_pvalue(r_all, n_all)
    r_c,   n_c   = pearson_r(xc, yc); p_c   = fisher_z_pvalue(r_c, n_c)
    r_g,   n_g   = pearson_r(xg, yg); p_g   = fisher_z_pvalue(r_g, n_g)

    tbl = pd.DataFrame({
        "Group": ["All","Control","Guidance"],
        "r": [r_all, r_c, r_g],
        "p (two-tailed)": [p_all, p_c, p_g],
        "n": [n_all, n_c, n_g]
    })
    with pd.option_context('display.float_format', lambda v: f"{v:.3f}"):
        print(f"[解释] {title}")
        display(tbl)

for xcol in construct_cols:
    if xcol not in merged.columns:
        continue
    scatter_and_table(
        merged, xcol, score_col,
        title=f"{xcol} vs Quiz Total Score (0–11)",
        xlabel=xcol, ylabel="Quiz Total Score (0–11)"
    )
```
